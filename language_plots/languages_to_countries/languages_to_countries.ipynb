```python
natural_earth_names = pd.read_csv("data/name_to_natural_name.csv")
natural_earth_names.rename(columns={"natural_name": "name",}, inplace=True)
natural_earth_names.columns
```

```python
countries_to_languages_map = defaultdict(list)

for index, row in countries_languages.iterrows():
    language_lst = ast.literal_eval(row["languages"])
    for language in language_lst:
        countries_to_languages_map[language].append(row["name"])
```

```python
for language, countries in countries_to_languages_map.items():
    for country in countries:
        # not the right natural earth name, but the right coloquial name
        if (country not in natural_earth_names["NAME"].values and
                country in natural_earth_names["NAME"].keys()):
            natural_earth_name = natural_earth_names["NAME"][country]
            countries_to_languages_map[natural_earth_name] = countries_to_languages_map.pop(country)           

with open("languages_to_countries.json", "w") as f:
    json.dump(countries_to_languages_map, f, indent=4)
```
